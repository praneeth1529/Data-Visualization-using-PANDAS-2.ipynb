```python

townName=[]
matchesplayed=[]
TownDate=df['home_team'].value_counts()
for _ in TownDate.index:
  if TownDate[_]>30:
   townName.append(_)
   matchesplayed.append(TownDate [_])
```

```python
from matplotlib import pyplot as plt
```

```python

plt.plot(townName,matchesplayed)
plt.ylabel('Matches Played')
plt.xlabel('Town Name')
plt.title('Matches Played by Towns')
plt.show()
```

```python

plt.bar(townName,matchesplayed)
plt.ylabel('Matches Played')
plt.xlabel('Town Name')
plt.title('Matches Played by Towns')
plt.show()
```

```python
plt.barh(townName,matchesplayed)
plt.ylabel('Matches Played')
plt.xlabel('Town Name')
plt.title('Matches Played by Towns')
plt.show()
```
